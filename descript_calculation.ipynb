PREPARING THE X AND Y DATA MATRICES 

X data matrix

```python
import subprocess
import os

# Paths to your .smi file and output file
smi_file = r'C:\Users\saman\Desktop\DESSERDATION\molecule.smi'
output_file = r'C:\Users\saman\Desktop\DESSERDATION\descriptors_output.csv'

# Path to the PaDEL-Descriptor .jar file
padel_jar = r"C:\Users\saman\PaDEL-Descriptor\PaDEL-Descriptor.jar"

# Run PaDEL-Descriptor using subprocess
command = [
    'java', '-Xms1G', '-Xmx1G', '-Djava.awt.headless=true', 
    '-jar', padel_jar, 
    '-i', smi_file, 
    '-o', output_file,
    '-fingerprints',  # Option to generate fingerprints
    '-standardizenitro',  # Standardize Nitro groups
    '-removesalt'  # Option to remove salts
]

try:
    # Run the command and capture the output
    process = subprocess.run(command, check=True, capture_output=True, text=True)

    # Print the output from the process (optional, for debugging)
    print(process.stdout)
    print("Descriptors generated successfully!")

except subprocess.CalledProcessError as e:
    print(f"Error occurred while running PaDEL-Descriptor: {e}")
    print(f"Error output: {e.stderr}")
```
